# Part One


You climb the hill and again try contacting the Elves. <br>However, you instead receive a signal you weren't expecting: a distress signal.

Your handheld device must still not be working properly; <br>the packets from the distress signal got decoded out of order. <br>You'll need to re-order the list of received packets (your puzzle input) to decode the message.
<br>
Your list consists of pairs of packets; pairs are separated by a blank line. <br>You need to identify how many pairs of packets are in the right order.

Packet data consists of lists and integers. Each list starts with [, ends with ], <br>and contains zero or more comma-separated values (either integers or other lists). <br>Each packet is always a list and appears on its own line.

When comparing two values, the first value is called left and the second value is called right. Then:<br>

If both values are integers, the lower integer should come first. If the left integer is lower than the right integer, <br>the inputs are in the right order. <br>If the left integer is higher than the right integer, the inputs are not in the right order. Otherwise, the inputs are the same integer; continue checking the next part of the input.<br>
If both values are lists, compare the first value of each list, then the second value, and so on. <br>If the left list runs out of items first, the inputs are in the right order. <br>If the right list runs out of items first, the inputs are not in the right order. If the lists are the same length and no comparison makes a decision about the order, continue checking the next part of the input.<br>
If exactly one value is an integer, convert the integer to a list which contains that integer as its only value, then retry the comparison. <br>For example, if comparing [0,0,0] and 2, <br>convert the right value to [2] (a list containing 2); the result is then found by instead comparing [0,0,0] and [2].<br>

What are the indices of the pairs that are already in the right order? (The first pair has index 1, the second pair has index 2, and so on.) In the above example, <br>the pairs in the right order are 1, 2, 4, and 6; the sum of these indices is 13.

Determine which pairs of packets are already in the right order. What is the sum of the indices of those pairs?<br>

```python
def compare(a, b, verbose=True):
    if isinstance(a, int) and isinstance(b, int):
        if verbose:
            print(f"{a} : {b} = {a<b}")
        if a < b:
            return -1
        elif b < a: 
            return 1
        elif a == b:
            return 0
        else:
            print(f"What am I doing here {a} : {b}")
    elif isinstance(a, list) and isinstance(b, list):
        res = 0
        for i in range(min(len(a), len(b))):
            res = compare(a[i], b[i], verbose)
            if res:
                break
        if res == 0:
            if len(a) < len(b):
                if verbose:
                    print(f"Left runs out")
                return -1
            elif len(b) < len(a):
                if verbose:
                    print(f"Right runs out")
                return 1
            else:
                if verbose:
                    print(f"List have same length!")
                return 0
    elif isinstance(a, list) and isinstance(b, int):
        if verbose:
            print(f"Made {b} a list")
        res = compare(a, [b], verbose=verbose)
    elif isinstance(a, int) and isinstance(b, list):
        if verbose:
            print(f"Made {a} a list")
        res = compare([a], b, verbose=verbose)
    else:
        print(f"Unexpected {a} and {b}")
    return res
```

Afterward, locate the divider packets. To find the decoder key for this distress signal, <br>you need to determine the indices of the two divider packets and multiply them together.<br> (The first packet is at index 1, the second packet is at index 2, and so on.) In this example, the divider packets are 10th and 14th, and so the decoder key is 140.<br>

```python
data = get_puzzle(year=2022, day=13)
data = data.split("\n")
data = [x for x in data if x != ""]
```

```python
with open(path, 'r') as file:
  # Read the contents of the file into a list
  data = file.readlines()
```

```python
data = [x.replace("\n", "") for x in data]
```

```python
data = [x for x in data if x != ""]
```

```python
results = deepcopy(data)
```

```python
dividers = ['[[2]]', '[[6]]']
results.extend(dividers)
```

```python
for i, d in enumerate(results):
    js = list(range(i))
    js.reverse()
    for j in js:
        if compare(eval(results[j]), eval(d), verbose=False) == 1:
            # print(f"Swapping {d} at {i} and {results[j]} at {j}")
            packet_d = results.pop(i)
            packet_dd = results.pop(j)
            results.insert(j, packet_d)
            results.insert(i, packet_dd)
            i = j
        else:
            break
results
```

```python
divider_idx = [i+1 for i, x in enumerate(results) if x in dividers]
print(divider_idx)
answer = divider_idx[0] * divider_idx[1]
print(answer)
```
